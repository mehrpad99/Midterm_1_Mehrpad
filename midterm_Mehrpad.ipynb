```python
def recommend_trails(park_data, park_name, min_elev=None, max_elev=None, min_rating=None, trail_type=None):
    if min_elev is None:
        min_elev = 0  
    
    recommended = park_data[park_data['area_name'] == park_name]
    if min_elev is not None:
        recommended = recommended[recommended['elevation_gain'] >= min_elev]
    if max_elev is not None:
        recommended = recommended[recommended['elevation_gain'] <= max_elev]
    if min_rating is not None:
        recommended = recommended[recommended['avg_rating'] >= min_rating]
    if trail_type is not None:
        recommended = recommended[recommended['route_type'] == trail_type]
    
    return recommended

recommend_trails(park_data, "Haleakala National Park", min_elev=1000, min_rating=4)
```

```python
def scrape_temps(month=None):
    my_url = "https://www.extremeweatherwatch.com/us-state-averages"
    response = requests.get(my_url)
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find("table", attrs={"class": "bordered-table sort-table"})
    rows = table.find_all('tr')
    headers = [header.text.strip() for header in rows[0].find_all('th')]
    month_index = headers.index(month.capitalize()) if month else headers.index('Average')
    temperatures = []
    for row in rows[1:]:
        cols = row.find_all('td')
        state = cols[0].text.strip()
        temp = cols[month_index].text.strip()
        temperatures.append([state, temp])
    temp_df = pd.DataFrame(temperatures, columns=['State', 'Temperature'])
    return temp_df
```

```python
import pandas as pd

recommended_trails = recommend_trails(
    park_data,
    min_elev=1000,
    max_elev=3000,
    min_rating=4.5,
    trail_type='loop'
)

# since scarping did not work properly (Code used is from writing functions cours on Datacamp for try, except, if and else)
try:
    temps_may = scrape_temps(month='may')
    temps_june = scrape_temps(month='june')
    temps_july = scrape_temps(month='july')
    temps_august = scrape_temps(month='august')
except Exception as e:
    print(f"Failed to scrape data due to: {str(e)}. Using CSV data instead.")
    temp_data = pd.read_csv('temp_data.csv') 

if 'temp_data' not in locals():
    summer_temps = pd.concat([temps_may, temps_june, temps_july, temps_august])
else:
    summer_temps = temp_data

merged_data = pd.merge(recommended_trails, summer_temps, on='state_name')

merged_data['elevation_feet'] = merged_data['elevation_gain'] * 3.28084
merged_data['adjusted_temp'] = merged_data['Temperature'] - (merged_data['elevation_feet'] / 1000) * 5.4

final_recommendations = merged_data[['name', 'area_name', 'state_name', 'adjusted_temp']]

final_recommendations
```
